```python
#INPUT
a,b,c=1,-5,6

#Process
D=b**2-4*a*c

if D>0:
  sifat_akar='real berbeda'
  x1=(-b+math.sqrt(D))/(2*a)
  x2=(-b-math.sqrt(D))/(2*a)
elif D==0:
  sifat_akar='akar kembar'
  x1=(-b+math.sqrt(D))/(2*a)
  x2=x1
else:
  sifat_akar='akar imajiner'

#Output
print(f"Nilai deskriminan(D)={D}")
print(f"Sifat akar adalah x1,x2 {sifat_akar}")

if sifat_akar!='akar imajiner':
  print(f"yaitu  x1={x1},x2= {x2}")
```
