```python
#Buat program menghitung pembayaran belanja
#dengan syarat jika total belanja >100K maka dapat diskon 5%

PPN=0.11  #====> ini konstanta

total_belanja=350000
dapat_diskon=total_belanja>500000

if dapat_diskon:
  diskon=total_belanja*0.05
  total_bayar=total_belanja-diskon
else:
  diskon=0
  total_bayar=total_belanja

ppn=total_bayar*PPN

print(f"Total Belanja ={total_belanja:,.2f} ")
print(f"Total Diskon  ={diskon:,.2f} ")
print(f"Total Bayar   ={total_bayar:,.2f} ")
print("="*27)
print(f"PPN           ={ppn:,.2f} ")
print(f"Bayar+PPN     ={total_bayar+ppn:,.2f} ")
```
